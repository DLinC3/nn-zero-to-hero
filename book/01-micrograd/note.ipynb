```python
class Value:
    
    def __init__(self, data, _children=(), _op=''):
        self.data = data
        # _prev stores the parent nodes that created this node
        self._prev = set(_children)
        self._op = _op
        
    def __repr__(self):
        return f"Value(data={self.data}, op={self._op})"
        
    def __add__(self, other):
        # Create a new Value and pass self and other as children
        out = Value(self.data + other.data, (self, other), '+')
        return out
        
    def __mul__(self, other):
        # Create a new Value and pass self and other as children
        out = Value(self.data * other.data, (self, other), '*')
        return out
```

```python
a = Value(2.0)
b = Value(-3.0)
c = Value(10.0)

# This performs: a.__mul__(b).__add__(c)
# a * b creates a temporary Value(data=-6.0) with children {a, b}
# d = a.__mul__(b).__add__(c)
d = a * b + c

print("d:", d)
print("d._prev:", d._prev)
```
